```python
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
plt.hist(gp["Close"])
plt.title("Price Distribution")
plt.xlabel("Price In Dollar($)")
plt.show()
```

```python
plt.plot(gp["Date"], gp["Close"])
plt.title("Gold Price Over Time")
plt.xlabel("Date")
plt.ylabel("Price")
plt.show()
```

```python
plt.scatter(gp["Open"], gp["Close"])
plt.xlabel("Opening Price")
plt.ylabel("Closing Price")
plt.show()
```

```python
gp.groupby("Year")["Close"].mean().plot(kind="bar")
plt.title("Average Gold Price by Year")
plt.show()
```
